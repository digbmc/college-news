## Changing Marker Type and Colors

### Marker Type

It helps convey more information in addition to being aesthetically pleasing.

```python
plt.style.use('dark_background')
plt.plot(month_number, principal_paid, marker = '.', markersize = 10)
plt.plot(month_number, interest_paid, marker = '.', markersize = 10)
```

### Changing Color

```python
plt.plot(month_number, principal_paid, c = 'y', marker = '.', markersize = 10)
plt.plot(month_number, interest_paid, c = 'b', marker = '.', markersize = 10)
```

The color pramater also accepts rgb tuples or strings denoting colors.

## MATLAB-style vs. Object Syntax

#### MATLAB-style:
- A scripted interface designed to feel like MATLAB where Matplotlib maintains a pointer to the active figure.

#### Object-oriented syntax
- Used to gain more control over the figure

- most plots are created using a combination of the two

### MATLAB-style
- example seen above
- plt.plot(x-variable, y-variable, parameter = " ")

### Object-oriented

```python
x,y = (3, 9)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(month_number, principal_paid, c = 'y')
axes.plot(month_number, interest_paid, c = 'b')
```

### Combination

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
plt.plot(month_number, principal_paid, c = 'y')
axes.plot(month_number, interest_paid, c = 'b')
```

## Titles, Lables, and Limits

### MATLAB-style

```python
plt.plot(month_number, principal_paid, c = 'y', marker = '.', markersize = 10)
plt.plot(month_number, interest_paid, c = 'b', marker = '.', markersize = 10)
```

#### Set x-limit and y-limit

```python
plt.plot(month_number, principal_paid, c = 'y', marker = '.', markersize = 10)
plt.plot(month_number, interest_paid, c = 'b', marker = '.', markersize = 10)
plt.xlim(left = 1, right = 70)
plt.ylim(bottom = 0, top = 1000)
```

#### Set x-label and y-label

```python
plt.plot(month_number, principal_paid, c = 'y', marker = '.', markersize = 10)
plt.plot(month_number, interest_paid, c = 'b', marker = '.', markersize = 10)
plt.xlabel('Month')
plt.ylabel('Dollars')
```

#### Set Title

```python
plt.plot(month_number, principal_paid, c = 'y', marker = '.', markersize = 10)
plt.plot(month_number, interest_paid, c = 'b', marker = '.', markersize = 10)
plt.xlabel('Month')
plt.ylabel('Dollars')
plt.title('Interest and Principal Paid Each Month')
```

#### Changing Fontsize

```python
plt.plot(month_number, principal_paid, c = 'y', marker = '.', markersize = 10)
plt.plot(month_number, interest_paid, c = 'b', marker = '.', markersize = 10)
plt.xlabel('Month', fontsize = 18)
plt.ylabel('Dollars', fontsize = 18)
plt.title('Interest and Principal Paid Each Month', fontsize = 22)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
```

### Object-oriented

#### Set x-limit and y-limit

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(month_number, principal_paid, c = 'y')
axes.plot(month_number, interest_paid, c = 'b')
axes.set_xlim(left = 1, right = 70)
axes.set_ylim(bottom = 0, top = 1000)
```

#### Set x-label and y-label

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(month_number, principal_paid, c = 'y')
axes.plot(month_number, interest_paid, c = 'b')
axes.set_xlabel('Month')
axes.set_ylabel('Dollars')
axes.set_title('Interest and Principal Paid Each Month')
```

#### Change Fontsize

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(month_number, principal_paid, c = 'y')
axes.plot(month_number, interest_paid, c = 'b')
axes.set_xlabel('Month', fontsize = 22)
axes.set_ylabel('Dollars', fontsize = 22)
axes.set_title('Interest and Principal Paid Each Month', fontsize = 22)
axes.tick_params(axis = 'x', labelsize = 20)
axes.tick_params(axis = 'y', labelsize = 20)
```

## Grids

### MATLAB-style

```python
plt.plot(month_number, principal_paid, c = 'y')
plt.plot(month_number, interest_paid, c = 'b')
plt.grid()
```

```python
plt.plot(month_number, principal_paid, c = 'y')
plt.plot(month_number, interest_paid, c = 'b')
plt.grid(axis = "x")
```

#### Change color, transparency, and linestyle

```python
plt.plot(month_number, principal_paid, c = 'y')
plt.plot(month_number, interest_paid, c = 'b')
plt.grid(axis = "x", 
        c = 'g',
        alpha = .5,
        linestyle = '-')
```

- alpha = 1: not transparent

## Plot Legends
- assign meaning to different elements in the plot

```python
plt.style.use('classic')
plt.plot(month_number, principal_paid, c = 'y', label = 'principal')
plt.plot(month_number, interest_paid, c = 'b', label = 'interest')
plt.legend()
```

```python
plt.style.use('classic')
plt.plot(month_number, principal_paid, c = 'y', label = 'principal')
plt.plot(month_number, interest_paid, c = 'b', label = 'interest')
plt.legend(loc=(1,0))
```

- loc changes the location of legends in the plot
    - e.g. center right or outside of the plot
    - object-oriented style follows similar syntax
